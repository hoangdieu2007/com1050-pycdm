```python
def GPA_check(lst1, lst2, lst3):
    if len(lst1) < 3 or len(lst2) < 2 or len(lst3) < 1:
        count = 0
        for v in lst1:
            count += v
        for n in lst2:
            count += n
        for m in lst3:
            count += m
        tbc = count / (len(lst1) + len(lst2) + len(lst3))
        if len(lst1) < 3:
            x = 3-len(lst1)
            lst1.extend([tbc]*x)
        if len(lst2) < 2:
            y = 2-len(lst2)
            lst2.extend([tbc]*y)
        if len(lst3) < 1:
            z = 1-len(lst3)
            lst3.extend([tbc]*z)
        count1 = 0
        count2 = 0
        count3 = 0
        for v in lst1:
            count1 += v
        for n in lst2:
            count2 += n
        for m in lst3:
            count3 += m
        Predicted_GPA = (count1/3)*(10/100) + (count2/2)*(30/100) + count3*(60/100)
        if Predicted_GPA >= 9.0:
            print(f"GPA Dự Đoán: {Predicted_GPA}\nHọc Lực: Xuất Sắc")
        elif 8.0 <= Predicted_GPA < 9.0:
            print(f"GPA Dự Đoán: {Predicted_GPA}\nHọc Lực: Giỏi")
        elif 7.0 <= Predicted_GPA < 8.0:
            print(f"GPA Dự Đoán: {Predicted_GPA}\nHọc Lực: Khá")
        elif 5.0 <= Predicted_GPA < 7.0:
            print(f"GPA Dự Đoán: {Predicted_GPA}\nHọc Lực: Trung Bình")
        else:
            print(f"GPA Dự Đoán: {Predicted_GPA}\nHọc Lực: Kém")
    elif len(lst1) > 3 or len(lst2) > 2 or len(lst3) > 1:
        print("Lỗi Nhập Liệu.")
    elif len(lst1) == 3 and len(lst2) == 2 and len(lst3) == 1:
        normal_score = 0
        midterm_score = 0
        final_score = 0
        for v in lst1:
            normal_score += v
        for n in lst2:
            midterm_score += n
        final_score = lst3[0]
        GPA = ((normal_score/3)*(10/100) + (midterm_score/2)*(30/100) + final_score*(60/100))
        if GPA >= 9.0:
            print(f"GPA: {GPA}\nHọc Lực: Xuất Sắc")
        elif 8.0 <= GPA < 9.0:
            print(f"GPA: {GPA}\nHọc Lực: Giỏi")
        elif 7.0 <= GPA < 8.0:
            print(f"GPA: {GPA}\nHọc Lực: Khá")
        elif 5.0 <= GPA < 7.0:
            print(f"GPA: {GPA}\nHọc Lực: Trung Bình")
        else:
            print(f"GPA: {GPA}\nHọc Lực: Kém")
lst1 = list(map(int, input().split()))
lst2 = list(map(int, input().split()))
lst3 = list(map(int, input().split()))
GPA_check(lst1, lst2, lst3)
    
```
